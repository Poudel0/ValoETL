```python
map_details =("Champions_Tour_2025_Pacific_Stage_1_-_Playoffs/double/89294/201633_details.json")
extra = ("Valorant_Champions_2024_-_Playoffs/double/80760/80760_extra.json")


with open(map_details, "r") as f:
    map_details = json.load(f)

with open(extra,'r') as f:
    extra = json.load(f)
```

```python
def print_tree(data, indent=0):
    spacing = "  " * indent
    if isinstance(data, dict):
        for key, value in data.items():
            print(f"{spacing}{key}/")
            print_tree(value, indent + 1)
    elif isinstance(data, list):
        print(f"{spacing}[list of {len(data)}]")
        if data:
            print_tree(data[0], indent + 1)  # Just show structure of first element
    else:
        print(f"{spacing}{type(data).__name__}")
```

```python
print_tree(map_details)
```

```python
print_tree(extra)
```
